```python
team_info = pd.read_csv('../input/team_info.csv', delimiter=',', usecols=['team_id', 'shortName', 'abbreviation'])
team_info.head()
```

```python
game_team_stats = pd.read_csv('../input/game_teams_stats.csv', delimiter=',', usecols=['game_id', 'team_id', 'HoA', 'won', 'settled_in', 'goals', 'shots', 'powerPlayGoals'])
game_team_stats = game_team_stats[game_team_stats.HoA == 'home']
game = pd.read_csv('../input/game.csv', delimiter=',', usecols=['game_id', 'season', 'home_goals', 'away_goals', 'home_team_id', 'away_team_id'])
print(game_team_stats.columns)
print(game.columns)
```

```python
final_game_results = game_team_stats.join(game.set_index('game_id'), on='game_id').sort_values(by=['game_id'])
final_game_results.head()
```
